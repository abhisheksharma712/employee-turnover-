```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/Mall_Customers.csv')
```

```python
df.head()
```

```python
x=df.iloc[:,[3,4]].values
```

```python
df.columns
```

```python
sns.scatterplot(data=df,x= 'Annual Income (k$)',y='Spending Score (1-100)',hue= 'Genre')
plt.show()
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 15):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x)  # Assuming x is your data
    wcss.append(kmeans.inertia_)
# kmeans.inertia_ is output of
plt.plot(range(1, 15), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=43)
y_kmeans=kmeans.fit_predict(x)
print(y_kmeans)
```

```python
plt.figure(figsize=(12,12))
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=50,color='red',label='Cluster1')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=50,color='purple',label='Cluster2')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=50,color='yellow',label='Cluster3')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=50,color='green',label='Cluster4')
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],s=50,color='black',label='Cluster5')
plt.legend()
plt.xlabel('Income')
plt.ylabel('Spending score')
plt.show()
```
