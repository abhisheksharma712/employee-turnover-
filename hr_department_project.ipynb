```python
 # separating categorical variables and numeric variables 
categorical_cols = df.select_dtypes(include=['object']).columns
numeric_cols = df.select_dtypes(exclude=['object']).columns

categorical_data = df[categorical_cols]
numeric_data = df[numeric_cols]
```

```python
categorical_data
```

```python
numeric_data 
```

```python
 # Applying get_dummies() to the categorical variables.
categorical_data_encoded=pd.get_dummies(categorical_data)
```

```python
categorical_data_encoded
```

```python
# Combining categorical variables and numeric variables.

processed_data = pd.concat([numeric_data, categorical_data_encoded],axis=1)
```

```python
processed_data
```

Do the stratified split of the dataset to train and test in the ratio 80:20 with random_state=123.

```python
from sklearn.model_selection import train_test_split

X = processed_data.drop('left', axis=1)
y = processed_data['left']

# Stratified split with 80:20 ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=123)
```

Upsample the train dataset using SMOTE technique from the imblearn module

```python
from imblearn.over_sampling import SMOTE

# Create SMOTE object
smote = SMOTE(random_state=123)

# Upsample only the training data
x_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
x1=x_train_resampled
```

```python
y1=y_train_resampled
```

```python
import pandas as pd
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt




def perform_cross_validation_and_plot(model, x1, y1, model_name):
    # Perform 5-Fold cross-validation
    y_pred = cross_val_predict(model, x1, y1, cv=5)
    
    # Print and plot classification report
    print(f"Classification Report for {model_name}:\n{classification_report(y1, y_pred)}")


logistic_regression_model = LogisticRegression()
perform_cross_validation_and_plot(logistic_regression_model, x1, y1, 'Logistic Regression')


random_forest_model = RandomForestClassifier()
perform_cross_validation_and_plot(random_forest_model, x1, y1, 'Random Forest Classifier')


gradient_boosting_model = GradientBoostingClassifier()
perform_cross_validation_and_plot(gradient_boosting_model, x1, y1, 'Gradient Boosting Classifier')


plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt


def calculate_roc_auc_and_plot(model, x1, y1, model_name):
    y_prob = cross_val_predict(model, x1, y1, cv=5, method='predict_proba')[:, 1]
    fpr, tpr, thresholds = roc_curve(y1, y_prob)
    auc = roc_auc_score(y1, y_prob)
    
    
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {auc:.2f})')


calculate_roc_auc_and_plot(logistic_regression_model, x1, y1, 'Logistic Regression')
calculate_roc_auc_and_plot(random_forest_model, x1, y1, 'Random Forest Classifier')
calculate_roc_auc_and_plot(gradient_boosting_model, x1, y1, 'Gradient Boosting Classifier')


plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve for Different Models')
plt.legend()
plt.show()
```
