```python
# now apply one hot encoding
df_train.head()
df_train_x=df_train.iloc[:,1:]
df_train_y=df_train.iloc[:,0]
df_test_x=df_test
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(handle_unknown='ignore',drop='first',sparse=False,dtype=np.int32)
df_train_x_new=ohe.fit_transform(df_train_x)
df_test_x_new=ohe.transform(df_test_x)
```

```python
# standarisation of data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_train_x_new=scaler.fit_transform(df_train_x_new)
df_test_x_new=scaler.transform(df_test_x_new)

```

```python
 # now apply dimensionality reduction via PCA
 from sklearn.decomposition import PCA
pca = PCA(n_components=100)
df_train_x_new = pca.fit_transform(df_train_x_new)
df_test_x_new = pca.transform(df_test_x_new)
```

```python
from xgboost import XGBRegressor

# Create XGBoost regressor object
xgb = XGBRegressor(n_estimators=100, max_depth=3, learning_rate=0.1, objective='reg:squarederror')

# Train the model
model = xgb.fit(df_train_x_new, df_train_y)

# Make predictions
y_pred = model.predict(df_test_x_new)
```
